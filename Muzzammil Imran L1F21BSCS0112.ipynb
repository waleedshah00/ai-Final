```python
#Muzzammil Imran L1F21BSCS0112
import random

#Muzzammil Imran L1F21BSCS0112
# Initialization
def initialize_population(population_size, chromosome_length):
    return [[random.randint(0, 1) for _ in range(chromosome_length)] for _ in range(population_size)]

#Muzzammil Imran L1F21BSCS0112
# Fitness Evaluation
def calculate_fitness(individual):
    return sum(individual)
    

#Muzzammil Imran L1F21BSCS0112
# Selection
def selection(population, fitness_values):
    return random.choices(population, weights=fitness_values, k=len(population))

#Muzzammil Imran L1F21BSCS0112
# Reproduction
def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

#Muzzammil Imran L1F21BSCS0112
# Mutation
def mutate(individual, mutation_rate):
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            individual[i] = 1 - individual[i]
    return individual

#Muzzammil Imran L1F21BSCS0112
# Replacement
def replace_population(population, offspring):
    combined_population = population + offspring
    combined_fitness = [calculate_fitness(individual) for individual in combined_population]
#Muzzammil Imran L1F21BSCS0112
    sorted_population = [x for _, x in sorted(zip(combined_fitness, combined_population), key=lambda pair: pair[0], reverse=True)]
    return sorted_population[:len(population)]

#Muzzammil Imran L1F21BSCS0112
# Termination condition
def check_termination(population, max_generations, generation):
    best_fitness = max(calculate_fitness(individual) for individual in population)
    return generation >= max_generations or best_fitness == len(population[0])

#Muzzammil Imran L1F21BSCS0112
# Genetic Algorithm
def genetic_algorithm(population_size, chromosome_length, mutation_rate, max_generations):
    population = initialize_population(population_size, chromosome_length)
    for generation in range(max_generations):
        fitness_values = [calculate_fitness(individual) for individual in population]
        selected_individuals = selection(population, fitness_values)
#Muzzammil Imran L1F21BSCS0112
        offspring = []
        for _ in range(population_size // 2):
            parent1, parent2 = random.choices(selected_individuals, k=2)
            child1, child2 = crossover(parent1, parent2)
            child1 = mutate(child1, mutation_rate)
            child2 = mutate(child2, mutation_rate)
            offspring.extend([child1, child2])
#Muzzammil Imran L1F21BSCS0112        
        population = replace_population(population, offspring)
        if check_termination(population, max_generations, generation):
            break

    best_individual = max(population, key=calculate_fitness)
    return best_individual

#Muzzammil Imran L1F21BSCS0112
# Example usage
population_size = 100
chromosome_length = 10
mutation_rate = 0.01
max_generations = 1000

#Muzzammil Imran L1F21BSCS0112
best_solution = genetic_algorithm(population_size, chromosome_length, mutation_rate, max_generations)
print("Best solution found:", best_solution)
```

```python
# Muzzammil Imran L1F21BSCS0112
import random

# Muzzammil Imran L1F21BSCS0112
# Define the parameters for the GA
POPULATION_SIZE = 100
GENOME_LENGTH = 10
MUTATION_RATE = 0.01
GENERATIONS = 100


# Muzzammil Imran L1F21BSCS0112
# Fitness function: sum of population of random genomes 
def fitness_function(genome):
    return sum(genome)

# Muzzammil Imran L1F21BSCS0112
# Initialization: generate a population of random genomes
def initialize_population():
    population = []
    for _ in range(POPULATION_SIZE):
        genome = [random.randint(0, 1) for _ in range(GENOME_LENGTH)]
        population.append(genome)
    return population   

# Muzzammil Imran L1F21BSCS0112
def selection(population, fitnesses):
    tournament_size = 5
    selected = []
    for _ in range(POPULATION_SIZE):
        tournament = random.sample(list(zip(population, fitnesses)), tournament_size)
        winner = max(tournament, key=lambda x: x[1])
        selected.append(winner[0])
    return selected


# Muzzammil Imran L1F21BSCS0112
# Crossover: single point crossover
def crossover(parent1, parent2):
    crossover_point = random.randint(1, GENOME_LENGTH - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2


# Muzzammil Imran L1F21BSCS0112
def mutate(genome):
    for i in range(GENOME_LENGTH):
        if random.random() < MUTATION_RATE:
            genome[i] = 1 if genome[i] == 0 else 0
    return genome 


# Muzzammil Imran L1F21BSCS0112 
# Main GA loop
def genetic_algorithm():
    # Initialize population
    population = initialize_population()

    for generation in range(GENERATIONS):
        # Evaluate fitness of the population
        fitnesses = [fitness_function(genome) for genome in population]

# Muzzammil Imran L1F21BSCS0112 
        # Selection
        selected_population = selection(population, fitnesses)

        # Reproduction
        next_generation = []
        for i in range(0, POPULATION_SIZE, 2):
            parent1 = selected_population[i]
            parent2 = selected_population[i+1]
            child1, child2 = crossover(parent1, parent2)
            next_generation.append(mutate(child1))
            next_generation.append(mutate(child2))

# Muzzammil Imran L1F21BSCS0112 
        # Replacement
        population = next_generation

        # Termination: check if a perfect solution is found
        best_fitness = max(fitnesses)
        if best_fitness == GENOME_LENGTH:
            print(f"Perfect solution found in generation {generation}")
            break 

# Muzzammil Imran L1F21BSCS0112 
    # Final evaluation
    best_genome = max(population, key=fitness_function)
    best_fitness = fitness_function(best_genome)
    print(f"Best genome: {best_genome}, Fitness: {best_fitness}")

# Muzzammil Imran L1F21BSCS0112
# Run the GA 
genetic_algorithm()
```
